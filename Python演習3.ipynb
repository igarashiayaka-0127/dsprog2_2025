### 問1

自然数を入力とし、入力された自然数が3の倍数(Ex. 3, 6, 9, ...)であるときは`hoge`、3を含む数(Ex. 3, 13, 23, ...)であるときは`huga`、両方の条件を満たす時は`piyo`を出力する関数`nabeatsu`を定義してください

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
6

# テストケース2
13

# テストケース3
33
```

````python
def nabeatsu(num:int) -> str:
    if num % 3 == 0 and '3' in str(num):
        return "piyo"
    elif '3' in str(num):
        return "huga"
    elif num % 3 == 0:
        return "hoge"
    
    
    return
````

````python
case_1 = 6
case_2 = 13
case_3 = 33

print('case_1:', nabeatsu(case_1))
print('case_2:', nabeatsu(case_2))
print('case_3:', nabeatsu(case_3))
````

### 問2

異なる自然数を含む配列とターゲットとなる自然数を入力として、配列に含まれる自然数を足し合わせてターゲットとなる自然数になる組み合わせを出力する関数`combination_sum`を定義してください

このとき、同じ自然数は何度でも利用してよいものとし、もし組み合わせが存在しない場合は空の配列を返してください

例えば、配列`[2, 3, 6, 7]`、ターゲット`7`を入力としたときの出力は`[[2,2,3],[7]]`になります

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
配列: [2, 3, 6, 7]
ターゲット: 14

# テストケース2
配列: [2, 3, 5]
ターゲット: 8

# テストケース3
配列: [2, 9, 11]
ターゲット: 100
```

````python
def combination_sum(candidates: list, target: int) -> list:
    # 全く同じ結果にならないようにする
    # set()→重複を排除
    # list()→リストに戻す
    # sorted()→ソートする
    split_candidates = sorted(list(set(candidates)))
    
    # 0からtargetまでの答えを保存する箱を作る
    # 足し算の結果をそれぞれの箱に保存していくイメージ
    # ターゲットの数だけ用意することでそこまで辿り着いた時に答えを取り出せる
    dp = [[] for _ in range(target + 1)]
    
    # 何もない時には空の配列を返す
    dp[0] = [[]]
    
    # 候補の数字を一つずつ使っていく
    for num in split_candidates:
        # 箱を一つずつ確認する(スタートは初めの数字から)
        for i in range(num, target + 1):
            # 組み合わせから答えを作る
            for prev_comb in dp[i - num]:
                dp[i].append(prev_comb + [num])
                
    # ターゲットの箱にたどり着いたら返す
    return dp[target]
````

````python
candidates_1 = [2, 3, 6, 7]
target_1 = 14

candidates_2 = [2, 3, 5]
target_2 = 8

candidates_3 = [2, 9, 11]
target_3 = 100

print(combination_sum(candidates_1, target_1))
print(combination_sum(candidates_2, target_2))
print(combination_sum(candidates_3, target_3))
````

### 問3

括弧を含む文字列を入力として、括弧の内包関係が正しいかどうかを判定し、正しければ`True`、正しくなければ`False`を返す関数`format_checker`を定義してください

例えば、`[](){}`や`[({})]`は正しい、`[(]`や`{[(})]`は正しくないと判定します

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
Hello I’m [(firstname) (lastname)] and I’m {age} years old!

# テストケース2
Hello I’m [(firstname) (lastname]) and I’m {age} years old!

# テストケース3
Hello I’m ((firstname) (lastname) and I’m {age} years old!
```

````python
# def format_checker(text:str) -> bool:
#     if '[]' in str(text) or '()' in str(text) or '{}' in str(text):
#         return True
#     elif text.count('[') != text.count(']'):
#         return False
#     elif text.count('(') != text.count(')'):
#         return False
#     elif text.count('{') != text.count('}'):
#         return False
#     return

def format_checker(text: str) -> bool:
    # 括弧のペアを辞書で定義（閉じ括弧がキー、開き括弧が値）
    pairs = {')': '(', ']': '[', '}': '{'}
    stack = []

    for char in text:
        # 足し算、引き算のイメージでその中で括弧の順番のルールがあるから、上から順番にチェックする必要がある
        # stack→1番後ろに追加、1番後ろから取り出す
        # 開き括弧なら、スタックに積む
        if char in '([{':
            stack.append(char)
            
        # 閉じ括弧なら、チェックを行う
        elif char in ')]}':
            # もしスタックが空なら、アウト
            if not stack:
                return False
            
            # スタックの一番上を取り出す
            last_open = stack.pop()
            
            # ペアが合っていなければ、アウト
            if last_open != pairs[char]:
                return False

    # ループが終わった後、スタックが空ならOK
    # 残っていれば、アウト
    if len(stack) == 0:
        return True
    else:
        return False
````

````python
case_1 = 'Hello I’m [(firstname) (lastname)] and I’m {age} years old!'
case_2 = 'Hello I’m [(firstname) (lastname]) and I’m {age} years old!'
case_3 = 'Hello I’m ((firstname) (lastname) and I’m {age} years old!'

print('case_1:', format_checker(case_1))
print('case_2:', format_checker(case_2))
print('case_3:', format_checker(case_3))
````
